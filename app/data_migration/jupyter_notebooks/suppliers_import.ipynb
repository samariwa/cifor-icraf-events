```python
import pandas as pd
df = pd.read_csv('../csv/suppliers.csv')
```

```python
headers = list(df.columns)
headers
```

```python
df = df.rename(columns={'Name': 'supplier_name', 'Supplier_contact': 'supplier_contact'})
```

```python
df
```

```python
df = df.fillna(0)
df = df.astype({'supplier_contact':'int'})
df = df.astype({'supplier_contact':'str'})
df['supplier_contact'] = df['supplier_contact'].str.zfill(10)
```

```python
df
```

```python
headers = list(df.columns)
headers.remove('Created_at')
headers.remove('Updated_at')
headers.remove('id')
headers
```

```python
for i, row in df.head(5).iterrows():
  values = {header: row[header] for header in headers}
  print(values)
```

```python
for i, row in df.iterrows():
  values = {header: row[header] for header in headers if row[header]}

  with open('../query_scripts/suppliers.py', 'a') as fp:
      fp.write('create_supplier(**'+str(values)+')\n')
```
