```python
df = df.rename(columns={'Number': 'phone_no', 
                        'Location': 'location',
                        'Note': 'note'})
```

```python
# new data frame with split value columns
split_name = df["Name"].str.split(" ", n = 1, expand = True)
 
# making separate first name column from new data frame
df["first_name"]= split_name[0]
 
# making separate last name column from new data frame
df["last_name"]= split_name[1]
```

```python
headers = list(df.columns)
headers.remove('Created_at')
headers.remove('Updated_at')
headers.remove('id')
headers.remove('User_id')
headers.remove('Name')
headers.remove('Deliverer')
headers.remove('Status')
headers
```

```python
for i, row in df.head(5).iterrows():
  values = {header: row[header] for header in headers}
  print(values)
```

```python
count = 2
for i, row in df.iterrows():
  values = {header: row[header] for header in headers if not pd.isna(row[header])}
  values['id'] = count
  count = count + 1
  if pd.isna(row['last_name']):
        values['last_name'] = ' '
  with open('../query_scripts/customers.py', 'a') as fp:
      fp.write('create_customer(**'+str(values)+')\n')
```
