```python
df = df.rename(columns={'Category_id': 'category_id',
                        'Name': 'product_name',
                        'Unit_id': 'unit_id',
                        'Subunit_id': 'sub_unit_id',
                        'Contains': 'contains',
                        'Quantity': 'total_qty',
                        'Restock_Level': 'restock_level'})
```

```python
headers = list(df.columns)
headers.remove('id')
headers.remove('Created_at')
headers.remove('Updated_at')
headers.remove('Supplier_id')
headers.remove('Buying_price')
headers.remove('Discount')
headers.remove('Opening_stock')
headers
```

```python
counter = 1
for i, row in df.iterrows():
  values = {}
  for header in headers:
    values['product_id'] = counter
    values['supplier_id'] = 1
    values['qty_purchased'] = row['total_qty']
    values['current_qty'] = row['total_qty']
    values['damaged'] = 0
    values['buying_price'] = row['Price'] / 2
    values['selling_discount'] = 0
    values['selling_price'] = row['Price']
    values['expiry_date'] = '2023-06-30'
  counter = counter + 1
  with open('../query_scripts/product_batches.py', 'a') as fp:
      fp.write('create_product_batch(**'+str(values)+')\n')
```
